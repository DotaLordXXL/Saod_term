### Задача 1. «Треугольник с максимальным периметром»
Массив A состоит из целых положительных чисел ­ длин отрезков. Составьте из трех отрезков такой треугольник, чтобы его периметр был максимально возможным. Если невозможно
составить треугольник с положительной площадью ­ функция возвращает 0.
### Ограничения:
• 3 ⩽ len(A) ⩽ 10000

• 1 ⩽ A[i] ⩽ 10

```python
%%time
import time
import random

def SelectionSort(A): 
    for i in range(0, len(A) - 1): 
        min_idx = i 
        for j in range(i + 1, len(A)): 
            if A[j] > A[min_idx]: 
                min_idx = j  
        A[i], A[min_idx] = A[min_idx], A[i]
        
def checkFunction(a, b, c):
    if a + b > c and a + c > b and b + c > a:
        return 1
    else:
        return 0
def perimeter(a, b, c):
    return a + b + c

import numpy as np
N = int(input("Введите длину массива от 3 до 10000: "))
if N < 3 or N > 10000:
    print("Число должно быть от 3 до 10000")
    N = int(input("Введите длину массива от 3 до 10000"))
A = np.empty(N)
for i in range(N):
        A[i] = random.randint(1, 10)
SelectionSort(A)
print(A)
perimeter_max = 0
for i in range(len(A) - 2):
        if checkFunction(A[i], A[i + 1], A[i + 2]):
            perimeter_max = perimeter(A[i],A[i + 1],A[i + 2])
            break
print(perimeter_max)
```

### Задача 2. «Максимальное число»
Дан массив неотрицательных целых чисел nums. Расположите их в таком порядке, чтобы
вместе они образовали максимально возможное число.
### Замечание
Результат может быть очень большим числом, поэтому представьте его как
string, а не integer
### Ограничения
1 ⩽ len(nums) ⩽ 100

• 0 ⩽ nums[i] ⩽ 109

```python
def maximum(num):
    num.sort(key=lambda x: x % 10, reverse = True)
    
    result = ''.join(str(i) for i in num)
    return result

print(maximum([3, 30, 34, 5, 9]))
```

### Задача 3. «Сортировка диагоналей в матрице»
Дана матрица mat размером m * n, значения ­ целочисленные. Напишите функцию, сортирующую каждую диагональ матрицы по возрастанию и возвращающую получившуюся матрицу.

### Ограничения
• m == len(mat)

• n == len(mat[i])

• 1 ⩽ m, n ⩽ 100

• 1 ⩽ mat[i][j] ⩽ 100


```python
import time
import random
def BubbleSort(A): 
    for j in range(len(A) - 1, 0, -1): 
        for i in range(0, j): 
            if A[i] > A[i + 1]: 
                A[i], A[i + 1] = A[i + 1], A[i]

import numpy as np
m = int(input("Введите количество строк от 3 до 10000: "))
if m < 1 or m > 100:
    print("Число должно быть от 1 до 100")
    m = int(input("Введите количество строк от 3 до 10000: "))
n = int(input("Введите количество столбцов от 3 до 10000: "))
if n < 1 or n > 100:
    print("Число должно быть от 1 до 100")
    m = int(input("Введите количество столбцов от 3 до 10000: "))
A = np.empty((m, n))
for i in range(m):
    for j in range(n):
        A[i, j] = random.randint(1, 100)
print(A)
print(" ")
B = np.empty(n)
for i in range(-(m-2), 1):
    np.resize(B, len(np.diag(A, k = i)))
    B = np.diagonal(A, offset = i).copy()
    BubbleSort(B)
    np.fill_diagonal(A[-i:], B)
for i in range(1, n):
    np.resize(B, len(np.diag(A, k = i)))
    B = np.diagonal(A, offset = i).copy()
    BubbleSort(B)
    np.fill_diagonal(A[:,i:], B)
print(A)
```

### Задача 4. «Объединение отрезков»
Дан массив отрезков intervals, в котором intervals[i] = [starti
, endi], некоторые
отрезки могут пересекаться. Напишите функцию, которая объединяет все пересекающиеся от­
резки в один и возвращает новый массив непересекающихся отрезков.
### Ограничения:
• 0 ⩽ len(intervals) ⩽ 104 

• len(intervals[i]) == 2

• 0 ⩽ starti < endi ⩽ 104

```python
def arr_Sort(A):
    for j in range(len(A) - 1, 0, -1): 
        for i in range(0, j): 
            if A[i][0] > A[i + 1][0]: 
                A[i], A[i + 1] = A[i + 1], A[i]

def unity(A):
    arr_Sort(A)
    print(A)
    B = [A[0]]
    j = 0
    for i in range(1, len(A)):
        if A[i][0] >= B[j][0] and A[i][0] <= B[j][1] and A[i][1] >= B[j][1]:
            B[j] = [B[j][0], A[i][1]]
        elif A[i][0] >= B[j][0] and A[i][0] <= B[j][1] and A[i][1] < B[j][1]:
            B[j] = [B[j][0], B[j][1]]
        else:
            B.append(A[i])
            j = j + 1
    return B
            
intervals = [[8,10], [1, 5], [3, 4], [2, 6], [15, 20], [20, 21]]
print(unity(intervals))
```

### Задача 5. «Шарики и стрелы»
Некоторые сферические шарики распределены по двухмерному пространству. Для каждого
шарика даны x­координаты начала и конца его горизонтального диаметра. Так как пространство
двумерно, то y­координаты не имеют значения в данной задаче. Координата xstart всегда меньше
xend.
Стрелу можно выстрелить строго вертикально (вдоль y­оси) из разных точек x­оси. Шарик
с координатами xstart и xend уничтожается стрелой, если она была выпущена из такой позиции
x, что xstart ⩽ x ⩽ xend. Когда стрела выпущена, она летит в пространстве бесконечное время
(уничтожая все шарики на пути).
Дан массив points, где points[i] = [xstart, xend]. Напишите функцию, возвращаю­
щую минимальное количество стрел, которые нужно выпустить, чтобы уничтожить все шари­
ки.

```python
def array_Sort(A):
    for j in range(len(A) - 1, 0, -1): 
        for i in range(0, j): 
            if A[i][0] > A[i + 1][0]: 
                A[i], A[i + 1] = A[i + 1], A[i]

def number(A):
    array_Sort(A)
    print(A)
    num = 1
    for i in range(len(A) - 1):
        for j in range(i + 1, len(A)):
            if A[j][0] > A[i][1]:
                num = num + 1
                break
    return num
            
points = [[10,16],[2,8],[1,6],[7,12]]
print(number(points))
points = [[1,2],[3,4],[5,6],[7,8]]
print(number(points))
points = [[1,2],[2,3],[3,4],[4,5]]
print(number(points))
points = [[1,2]]
print(number(points))
points = [[2,3],[2,3]]
print(number(points))
```

### Задача 6
Даны две строки: s1 и s2 с одинаковым размером, проверьте, может ли некоторая перестановка строки s1 “победить” некоторую перестановку строки s2 или наоборот. Строка x может “победить” строку y (обе имеют размер n), если x[i]> = y [i] (в алфавитном порядке) для всех i от 0 до n-1.

```python
def check(s1, s2):
    s1 = sorted(s1)
    s2 = sorted(s2)
    flag1 = True
    flag2 = True
    for i in range(len(s1)):
        if s1[i] > s2[i]:
            flag2 = False
        if s2[i] > s1[i]:
            flag1 = False
    return flag1 or flag2

s1 = "abc"
s2 = "cda"
print(check(s1, s2))
```

### Задача 7
Дана строка s, вернуть самую длинную полиндромную подстроку в s

```python
def Palindromes(s):
    s_rev = s[::-1]
    return s_rev == s
def Longest_Palindrome(s):
    max_len = 1
    Max_pal = s[0]
    for i in range(len(s) - 1):
        for j in range (i+1, len(s)):
            if Palindromes(s[i:j]) and len(s[i:j]) > max_len:
                max_len = len(s[i:j])
                Max_pal = s[i:j]
    return Max_pal

s = "abb2x2xr"
print(Longest_Palindrome(s))
```

### Задача 8
Вернуть количество отдельных непустых подстрок текста, которые могут быть записаны как конкатенация некоторой строки с самой собой (т.е. она может быть записана, как a + a, где a - некоторая строка).

```python
def Conc(s):
    s_arr = []
    for i in range(len(s) - 1):
        for j in range(i + 1, len(s)):
            if s.startswith(s[i:j], j) and not s[i:j] in s_arr:
                s_arr.append(s[i:j])
    return len(s_arr)

s = "abcabcabc"
print(Conc(s))
```

### Задача 9. «Стопки монет»
На столе стоят 3n стопок монет. Вы и ваши друзья Алиса и Боб забираете стопки монет по
следующему алгоритму:
1. Вы выбираете 3 стопки монет из оставшихся на столе.
2. Алиса забирает себе стопку с максимальным количеством монет.
3. Вы забираете одну из двух оставшихся стопок.
4. Боб забирает последнюю стопку.
5. Если еще остались стопки, то действия повторяются с первого шага.

Дан массив целых положительных чисел piles. Напишите функцию, возвращающую максимальное число монет, которое вы можете получить


```python
def game(piles):
    piles.sort()
    result = 0
    begin = len(piles) // 3
    for i in range(begin, len(piles), 2):
        result = result + piles[i]
    return result

piles = [2,4,1,2,7,8]
print(game(piles))
```
